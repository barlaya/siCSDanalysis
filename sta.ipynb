```python
N = len(vic)
id = 0
idx = list()
t = list()
mn = list()

#simítás:
def mov_wav(x):
  x[0] = x[0]
  x[1] = x[1]
  for i in range(2, len(x)-3):
    x[i] = (x[i-2] + x[i-1] + x[i] + x[i+1] + x[i+2])/5
  return x
avvic = mov_wav(list(vic))

#threshold:
for i in range(len(avvic[1:N])):
  if ((avvic[i-1]) < 0) & (avvic[i]>0):
    id = i
    idx.append(id)
idx = np.array(idx)


#no. of spikes:
print(len(idx), "db spike")

#time window on the signal
def spl(chn):
  '''
  gives a list of action potential spikes
  input: sign of a channel, on which I want to window the spikes
  the method copies time intervals defined by threshold indices
  the first and last spikes are unnecessary
  output: returns a t array containing "useful" voltage values
  '''
  t = list()
  for i in range(len(idx)):
    t.append(chn[idx[i]-100: idx[i]+100])
  t.pop(0)
  t.pop(-1)
  t.pop(653)
  t = np.array(t)
  return t
t = spl(vic)

#average IC spike:
mn = t.mean(axis=0)
spec = list() 
mnspec = list() 
for i in range(len(vec)):
  spec.append(spl(vec[i]))
  mnspec.append(spec[i].mean(axis=0))
```
